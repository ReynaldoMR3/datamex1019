```python
from selenium import webdriver
```

```python
navegador = webdriver.Chrome()
```

```python
navegador.get('https://www.overbuff.com/') #es importante usar siempre https
```

```python
navegador.find_element_by_xpath('/html/body/div[1]/div[1]/div[1]/div/div[2]/a').click()
```

```python
tabla = navegador.find_element_by_tag_name('table')
```

```python
tabla_encabezado = tabla.find_element_by_tag_name('thead')
```

```python
columnas = tabla_encabezado.find_element_by_tag_name('tr')
```

```python
tbody = navegador.find_element_by_tag_name('tbody')
```

```python
tr_body = navegador.find_elements_by_tag_name('tr')
```

```python
tr = [i.text for i in tr_body]
tr_complete = tr[1:]
```

```python
tr_clean = [i.split('\n') for i in tr_complete]
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(tr_clean, columns=(['Hero', 'Position', 'Pick_Rate', 'Win_Rate', 'Tie_Rate', 'On_Fire']))
```
